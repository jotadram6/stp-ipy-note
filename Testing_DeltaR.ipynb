```python
import ROOT
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Eta1=[]; Phi1=[]; Eta2=[]; Phi2=[]
DEta=[]; DPhi=[]
for i in xrange(100000):
    V=np.random.normal(loc=0.0, scale=5.0, size=4)
    LV1=ROOT.TLorentzVector(V[0],V[1],V[2],V[3])
    LV2=ROOT.TLorentzVector(0.,0.,0.,0.)
    LV2.SetPxPyPzE(-1.0*LV1.Px(),-1.0*LV1.Py(),-1.0*LV1.Pz(),LV1.E())
    Eta1.append(LV1.Eta()); Eta2.append(LV2.Eta())
    Phi1.append(LV1.Phi()); Phi2.append(LV2.Phi())
    DEta.append((LV1.Eta()-LV2.Eta())**2); DPhi.append(LV1.DeltaPhi(LV2)**2)
    if LV1.DeltaR(LV2)<3: print LV1.DeltaR(LV2)
```

```python
fig, ax = plt.subplots()
ax.plot(Eta1, Phi1, 'k.', markersize=2.0)
ax.plot(Eta2, Phi2, 'bx', markersize=2.0)
ax.set_xlabel("$\eta$")
ax.set_ylabel("$\phi$")
ax.xaxis.label.set_fontsize(18)
ax.yaxis.label.set_fontsize(18)
ax.set_xlim(left=-6.0,right=6.0)
ax.set_ylim(bottom=-3.14,top=3.14)
#plt.legend(loc='lower right')
#fig.savefig("Testing_etaphi_random_backtoback.pdf")

fig1, ax1 = plt.subplots()
ax1.plot(DEta, DPhi, 'k.', markersize=10.0)
ax1.set_xlabel("$(\Delta\eta)^{2}$")
ax1.set_ylabel("$(\Delta\phi)^{2}$")
ax1.xaxis.label.set_fontsize(18)
ax1.yaxis.label.set_fontsize(18)
ax1.set_xlim(left=0.0,right=100.0)
ax1.set_ylim(bottom=9.5,top=10.0)
```
