```python
from Generic import *
NominalV="V8"

SignalSampleFile="Arrays_V8V9V10/Version8/SignalSample_2ndTopMass_NCSVM_"+NominalV
MCSignalSampleArray=np.load(SignalSampleFile+".npy")
NominalNMassPointsA=MCSignalSampleArray[0:NsignalFiles]
del(MCSignalSampleArray)
```

```python
CutsFromCLs=[CutHT,CutDRbb,CutDRWH,CutHM,CutM2HP,CutDRTp6thJ,CutRelHT]
FullSel=CutHT*CutDRbb*CutDRWH*CutHM*CutM2HP*CutDRTp6thJ*CutRelHT
BinsLimits="(40,400,1600)"
Masses=[600,650,700,750,800,850,900,950,1000]
FitWindows=[100,100,150,150,150,150,150,150,150]

SF=ROOT.TCut("weight")
PU=ROOT.TCut("PUR_function(NTI)")
CMC=SF*PU
```

# PU Systematics

```python
PUUXS=ROOT.TCut("PUR_Ufunction(NTI)")
PUDXS=ROOT.TCut("PUR_Dfunction(NTI)")
CMCPUUXS=SF*PUUXS
CMCPUDXS=SF*PUDXS

PUUNTI=ROOT.TCut("PUR_function(round(NTI*1.05))")
PUDNTI=ROOT.TCut("PUR_function(round(NTI*0.95))")
CMCPUUNTI=SF*PUUNTI
CMCPUDNTI=SF*PUDNTI

NominalWInt=[]; NominalTInt=[]
UpXSWInt=[]; UpXSTInt=[]
DownXSWInt=[]; DownXSTInt=[]
UpNTIWInt=[]; UpNTITInt=[]
DownNTIWInt=[]; DownNTITInt=[]

PDFNameScan="PU_Systematics.pdf"

for i in xrange(len(Masses)):
    print "#########################"
    print "###### Mass =", Masses[i], "#######"
    print "#########################"
    print "----------------Nominal resulst-----------------------------"
    NominalNMassPointsA[i].Draw("M5J >> M5jS"+str(Masses[i])+"_CutF_N"+BinsLimits,FullSel*CMC)
    M5JHSN=ROOT.gDirectory.Get("M5jS"+str(Masses[i])+"_CutF_N")
    M5JHSN.Sumw2(); M5JHSN.Scale(NWeights[i])
    
    f1=ROOT.TF1("f1","gaus",Masses[i]-FitWindows[i],Masses[i]+FitWindows[i])
    M5JHSN.Fit("f1","LMR","same")
    MassPoint=f1.GetParameter(1); win=f1.GetParameter(2)
    MassPointE=f1.GetParError(1); winE=f1.GetParError(2)
    Chi2=f1.GetChisquare(); NDF=f1.GetNDF()
    window_error=ROOT.Double(0.)
    integral=M5JHSN.IntegralAndError(M5JHSN.GetXaxis().FindBin(MassPoint-win),M5JHSN.GetXaxis().FindBin(MassPoint+win),window_error)
    print "M="+str(Masses[i])+" GeV ------> {0:.2f}+-{1:.2f}".format(integral,window_error)
    print "Mean value: {0:.2f}+-{1:.2f}, with sigma: {2:.2f}+-{3:.2f}".format(MassPoint,MassPointE,win,winE)
    print "Chi2={0:.2f}, ndf={1:.2f}, Chi2/ndf={2:.2f}".format(Chi2, NDF, Chi2/NDF)
    NominalWInt.append(integral)
    NominalTInt.append(M5JHSN.Integral())
    
    print "----------------UP(XS) resulst-----------------------------"
    NominalNMassPointsA[i].Draw("M5J >> M5jS"+str(Masses[i])+"_CutF_UXS"+BinsLimits,FullSel*CMCPUUXS)
    M5JHSUXS=ROOT.gDirectory.Get("M5jS"+str(Masses[i])+"_CutF_UXS")
    M5JHSUXS.Sumw2(); M5JHSUXS.Scale(NWeights[i])
    
    f1=ROOT.TF1("f1","gaus",Masses[i]-FitWindows[i],Masses[i]+FitWindows[i])
    M5JHSUXS.Fit("f1","LMR","same")
    MassPoint=f1.GetParameter(1); win=f1.GetParameter(2)
    MassPointE=f1.GetParError(1); winE=f1.GetParError(2)
    Chi2=f1.GetChisquare(); NDF=f1.GetNDF()
    window_error=ROOT.Double(0.)
    integral=M5JHSUXS.IntegralAndError(M5JHSUXS.GetXaxis().FindBin(MassPoint-win),M5JHSUXS.GetXaxis().FindBin(MassPoint+win),window_error)
    print "M="+str(Masses[i])+" GeV ------> {0:.2f}+-{1:.2f}".format(integral,window_error)
    print "Mean value: {0:.2f}+-{1:.2f}, with sigma: {2:.2f}+-{3:.2f}".format(MassPoint,MassPointE,win,winE)
    print "Chi2={0:.2f}, ndf={1:.2f}, Chi2/ndf={2:.2f}".format(Chi2, NDF, Chi2/NDF)
    UpXSWInt.append(integral)
    UpXSTInt.append(M5JHSUXS.Integral())
    
    print "----------------DOWN(XS) resulst-----------------------------"
    NominalNMassPointsA[i].Draw("M5J >> M5jS"+str(Masses[i])+"_CutF_DXS"+BinsLimits,FullSel*CMCPUDXS)
    M5JHSDXS=ROOT.gDirectory.Get("M5jS"+str(Masses[i])+"_CutF_DXS")
    M5JHSDXS.Sumw2(); M5JHSDXS.Scale(NWeights[i])
    
    f1=ROOT.TF1("f1","gaus",Masses[i]-FitWindows[i],Masses[i]+FitWindows[i])
    M5JHSDXS.Fit("f1","LMR","same")
    MassPoint=f1.GetParameter(1); win=f1.GetParameter(2)
    MassPointE=f1.GetParError(1); winE=f1.GetParError(2)
    Chi2=f1.GetChisquare(); NDF=f1.GetNDF()
    window_error=ROOT.Double(0.)
    integral=M5JHSDXS.IntegralAndError(M5JHSDXS.GetXaxis().FindBin(MassPoint-win),M5JHSDXS.GetXaxis().FindBin(MassPoint+win),window_error)
    print "M="+str(Masses[i])+" GeV ------> {0:.2f}+-{1:.2f}".format(integral,window_error)
    print "Mean value: {0:.2f}+-{1:.2f}, with sigma: {2:.2f}+-{3:.2f}".format(MassPoint,MassPointE,win,winE)
    print "Chi2={0:.2f}, ndf={1:.2f}, Chi2/ndf={2:.2f}".format(Chi2, NDF, Chi2/NDF)
    DownXSWInt.append(integral)
    DownXSTInt.append(M5JHSDXS.Integral())
    
    print "----------------UP(NTI) resulst-----------------------------"
    NominalNMassPointsA[i].Draw("M5J >> M5jS"+str(Masses[i])+"_CutF_UNTI"+BinsLimits,FullSel*CMCPUUNTI)
    M5JHSUNTI=ROOT.gDirectory.Get("M5jS"+str(Masses[i])+"_CutF_UNTI")
    M5JHSUNTI.Sumw2(); M5JHSUNTI.Scale(NWeights[i])
    
    f1=ROOT.TF1("f1","gaus",Masses[i]-FitWindows[i],Masses[i]+FitWindows[i])
    M5JHSUNTI.Fit("f1","LMR","same")
    MassPoint=f1.GetParameter(1); win=f1.GetParameter(2)
    MassPointE=f1.GetParError(1); winE=f1.GetParError(2)
    Chi2=f1.GetChisquare(); NDF=f1.GetNDF()
    window_error=ROOT.Double(0.)
    integral=M5JHSUNTI.IntegralAndError(M5JHSUNTI.GetXaxis().FindBin(MassPoint-win),M5JHSUNTI.GetXaxis().FindBin(MassPoint+win),window_error)
    print "M="+str(Masses[i])+" GeV ------> {0:.2f}+-{1:.2f}".format(integral,window_error)
    print "Mean value: {0:.2f}+-{1:.2f}, with sigma: {2:.2f}+-{3:.2f}".format(MassPoint,MassPointE,win,winE)
    print "Chi2={0:.2f}, ndf={1:.2f}, Chi2/ndf={2:.2f}".format(Chi2, NDF, Chi2/NDF)
    UpNTIWInt.append(integral)
    UpNTITInt.append(M5JHSUNTI.Integral())
    
    print "----------------DOWN(NTI) resulst-----------------------------"
    NominalNMassPointsA[i].Draw("M5J >> M5jS"+str(Masses[i])+"_CutF_DNTI"+BinsLimits,FullSel*CMCPUDNTI)
    M5JHSDNTI=ROOT.gDirectory.Get("M5jS"+str(Masses[i])+"_CutF_DNTI")
    M5JHSDNTI.Sumw2(); M5JHSDNTI.Scale(NWeights[i])
    
    f1=ROOT.TF1("f1","gaus",Masses[i]-FitWindows[i],Masses[i]+FitWindows[i])
    M5JHSDNTI.Fit("f1","LMR","same")
    MassPoint=f1.GetParameter(1); win=f1.GetParameter(2)
    MassPointE=f1.GetParError(1); winE=f1.GetParError(2)
    Chi2=f1.GetChisquare(); NDF=f1.GetNDF()
    window_error=ROOT.Double(0.)
    integral=M5JHSDNTI.IntegralAndError(M5JHSDNTI.GetXaxis().FindBin(MassPoint-win),M5JHSDNTI.GetXaxis().FindBin(MassPoint+win),window_error)
    print "M="+str(Masses[i])+" GeV ------> {0:.2f}+-{1:.2f}".format(integral,window_error)
    print "Mean value: {0:.2f}+-{1:.2f}, with sigma: {2:.2f}+-{3:.2f}".format(MassPoint,MassPointE,win,winE)
    print "Chi2={0:.2f}, ndf={1:.2f}, Chi2/ndf={2:.2f}".format(Chi2, NDF, Chi2/NDF)
    DownNTIWInt.append(integral)
    DownNTITInt.append(M5JHSDNTI.Integral())
    
    print "----------------------------------------------------------"
    print "Systematics from 1-sigma integral window"
    print "XS Up Variation={0:.2f}".format(100*abs(UpXSWInt[-1]-NominalWInt[-1])/NominalWInt[-1])
    print "XS Down Variation={0:.2f}".format(100*abs(DownXSWInt[-1]-NominalWInt[-1])/NominalWInt[-1])
    print "NTI Up Variation={0:.2f}".format(100*abs(UpNTIWInt[-1]-NominalWInt[-1])/NominalWInt[-1])
    print "NTI Down Variation={0:.2f}".format(100*abs(DownNTIWInt[-1]-NominalWInt[-1])/NominalWInt[-1])
    print "Systematics from full integral"
    print "XS Up Variation={0:.2f}".format(100*abs(UpXSTInt[-1]-NominalTInt[-1])/NominalTInt[-1])
    print "XS Down Variation={0:.2f}".format(100*abs(DownXSTInt[-1]-NominalTInt[-1])/NominalTInt[-1])
    print "NTI Up Variation={0:.2f}".format(100*abs(UpNTITInt[-1]-NominalTInt[-1])/NominalTInt[-1])
    print "NTI Down Variation={0:.2f}".format(100*abs(DownNTITInt[-1]-NominalTInt[-1])/NominalTInt[-1])
    
    CurCanv = rootnotes.canvas("MyPlot", (600, 800))
    M5JHSN.SetStats(ROOT.kFALSE)
    M5JHSN.SetLineStyle(1); M5JHSUXS.SetLineStyle(9); M5JHSDXS.SetLineStyle(7); M5JHSUNTI.SetLineStyle(9); M5JHSDNTI.SetLineStyle(7)
    M5JHSN.SetLineWidth(4); M5JHSUXS.SetLineWidth(2); M5JHSDXS.SetLineWidth(2); M5JHSUNTI.SetLineWidth(2); M5JHSDNTI.SetLineWidth(2)
    
    M5JHSN.SetLineColor(ROOT.kGreen+1); M5JHSUXS.SetLineColor(ROOT.kRed); M5JHSDXS.SetLineColor(ROOT.kBlue); M5JHSUNTI.SetLineColor(ROOT.kRed); M5JHSDNTI.SetLineColor(ROOT.kBlue); 
    
    LEG1=ROOT.TLegend(0.6,0.6,0.9,0.9)
    LEG1.AddEntry(M5JHSN, "Nominal", "l")#; LEG.AddEntry(ROOT.TObject(), GetMR(M5JHQ), ""); LEG.AddEntry(ROOT.TObject(), GetEWI(M5JHQ), "")
    LEG1.AddEntry(M5JHSUXS, "Up", "l")#; LEG.AddEntry(ROOT.TObject(), GetMR(M5JHT), ""); LEG.AddEntry(ROOT.TObject(), GetEWI(M5JHT), "")
    LEG1.AddEntry(M5JHSDXS, "Down", "l")#; LEG.AddEntry(ROOT.TObject(), GetMR(M5JHS), ""); LEG.AddEntry(ROOT.TObject(), GetEWI(M5JHS), "")
    LEG1.SetFillColor(0)
    LEG2=ROOT.TLegend(0.6,0.6,0.9,0.9)
    LEG2.AddEntry(M5JHSN, "Nominal", "l")#; LEG.AddEntry(ROOT.TObject(), GetMR(M5JHQ), ""); LEG.AddEntry(ROOT.TObject(), GetEWI(M5JHQ), "")
    LEG2.AddEntry(M5JHSUNTI, "Up", "l")#; LEG.AddEntry(ROOT.TObject(), GetMR(M5JHT), ""); LEG.AddEntry(ROOT.TObject(), GetEWI(M5JHT), "")
    LEG2.AddEntry(M5JHSDNTI, "Down", "l")#; LEG.AddEntry(ROOT.TObject(), GetMR(M5JHS), ""); LEG.AddEntry(ROOT.TObject(), GetEWI(M5JHS), "")
    LEG2.SetFillColor(0)
    
    MinX=float(BinsLimits.split(",")[1])
    MaxX=float(BinsLimits.split(",")[-1][0:-1])
    Nbins=int(BinsLimits.split(",")[0][1:])
    M5JHSN.SetTitle(";M_{{5j}};Events/{0:.1f} GeV/c^{{2}}".format((MaxX-MinX)/Nbins))
    M5JHSN.GetYaxis().SetTitleOffset(1.5)
    
    Maxima1=[M5JHSN.GetMaximum(), M5JHSUXS.GetMaximum(), M5JHSDXS.GetMaximum()]
    Maxima2=[M5JHSN.GetMaximum(), M5JHSUNTI.GetMaximum(), M5JHSDNTI.GetMaximum()]
    
    for j in xrange(2):
        if j==0:
            M5JHSN.SetMaximum(1.05*max(Maxima1))
            M5JHSN.Draw("hist"); M5JHSUXS.Draw("hist same"); M5JHSDXS.Draw("hist same"); LEG1.Draw()
            if i==0: CurCanv.Print(PDFNameScan+"(","Title: (XS) M"+str(Masses[i]))
            else: CurCanv.Print(PDFNameScan,"Title: (XS) M"+str(Masses[i]))
        if j==1:
            M5JHSN.SetMaximum(1.05*max(Maxima2))
            M5JHSN.Draw("hist"); M5JHSUNTI.Draw("hist same"); M5JHSDNTI.Draw("hist same"); LEG2.Draw()
            if i==0: CurCanv.Print(PDFNameScan,"Title: (NTI) M"+str(Masses[i]))
            elif i==(len(Masses)-1): CurCanv.Print(PDFNameScan+")","Title: (NTI) M"+str(Masses[i]))
            else: CurCanv.Print(PDFNameScan,"Title: (NTI) M"+str(Masses[i]))
    
    del(CurCanv)
    
ROOT.gDirectory.Delete("*CutF*")
```
