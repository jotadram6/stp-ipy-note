# Signal fitting results

```python
#Plotting signal fitting results
MeanValues=[]
MeanErrors=[]
SigmaValues=[]
SigmaErrors=[]

for i in xrange(len(Masses)):
    print "#########################"
    print "###### Mass =", Masses[i], "#######"
    print "#########################"
    print "----------------Nominal resulst-----------------------------"
    NominalNMassPointsA[i].Draw("M5J >> M5jS"+str(Masses[i])+"_CutF_N"+BinsLimits,FullSel*CMC)
    M5JHSN=ROOT.gDirectory.Get("M5jS"+str(Masses[i])+"_CutF_N")
    #M5JHSN=AddingReWTriggerSys(FullSel*CMC,"M5J",BinsLimits,NominalNMassPointsA[i],"M5jS"+str(Masses[i])+"_CutF_N")
    M5JHSN.Sumw2(); M5JHSN.Scale(NWeights[i])
    
    f1=ROOT.TF1("f1","gaus",Masses[i]-FitWindows[i],Masses[i]+FitWindows[i])
    M5JHSN.Fit("f1","LMR","same")
    MassPoint=f1.GetParameter(1); win=f1.GetParameter(2)
    MeanValues.append(MassPoint); SigmaValues.append(win)
    MassPointE=f1.GetParError(1); winE=f1.GetParError(2)
    MeanErrors.append(MassPointE); SigmaErrors.append(winE)
    Chi2=f1.GetChisquare(); NDF=f1.GetNDF()
    window_error=ROOT.Double(0.)
    integral=M5JHSN.IntegralAndError(M5JHSN.GetXaxis().FindBin(MassPoint-win),M5JHSN.GetXaxis().FindBin(MassPoint+win),window_error)
    print "M="+str(Masses[i])+" GeV ------> {0:.2f}+-{1:.2f}".format(integral,window_error)
    print "Mean value: {0:.2f}+-{1:.2f}, with sigma: {2:.2f}+-{3:.2f}".format(MassPoint,MassPointE,win,winE)
    print "Chi2={0:.2f}, ndf={1:.2f}, Chi2/ndf={2:.2f}".format(Chi2, NDF, Chi2/NDF)
    
print "############### Fitting linearly masses and widths ##################"
from linfit import linfit
fit1, cvm, info = linfit(Masses, MeanValues, sigmay=SigmaValues, relsigma=False, return_all=True)
dfit1 = [np.sqrt(cvm[i,i]) for i in range(2)]
print "Masses fit:"
print(u"slope = {0:0.1f} \xb1 {1:0.1f}".format(fit1[0], dfit1[0]))
print(u"y-intercept = {0:0.0f} \xb1 {1:0.0f}".format(fit1[1], dfit1[1]))

fit2, cvm, info = linfit(Masses, SigmaValues, sigmay=SigmaValues, relsigma=False, return_all=True)
dfit2 = [np.sqrt(cvm[i,i]) for i in range(2)]
print "Widths fit:"
print(u"slope = {0:0.1f} \xb1 {1:0.1f}".format(fit2[0], dfit2[0]))
print(u"y-intercept = {0:0.0f} \xb1 {1:0.0f}".format(fit2[1], dfit2[1]))
print "#####################################################################"

fig, ax = plt.subplots()
ax.errorbar(Masses, MeanValues, yerr=SigmaValues, ecolor='k', marker='o', ms=5, color='black', ls='none')
endx = 0.001 * (max(Masses)-min(Masses))
xFit = np.array([min(Masses)-endx, max(Masses)+endx])
yFit = fit1[0]*xFit + fit1[1]
ax.plot(xFit, yFit, "-b", zorder=-1)
ax.set_xlabel("M(T) GeV/$c^{2}$")
ax.set_ylabel("Fitted M(5j)")
ax.xaxis.label.set_fontsize(18)
ax.yaxis.label.set_fontsize(18)
ax.xaxis.set_label_coords(0.84, -0.07)
ax.yaxis.set_label_coords(-0.09, 0.78)
ax.set_xlim(left=Masses[0]*0.95,right=Masses[-1]*1.05)
ax.set_ylim(bottom=min(MeanValues)*0.85,top=max(MeanValues)*1.15)
#plt.legend(loc='lower right')
fig.savefig("Fitted_TpMass_V8.pdf")
fig.savefig("Fitted_TpMass_V8.png")

fig, ax = plt.subplots()
ax.errorbar(Masses, SigmaValues, yerr=SigmaErrors, ecolor='k', marker='o', ms=5, color='black', ls='none')
endx = 0.001 * (max(Masses)-min(Masses))
xFit = np.array([min(Masses)-endx, max(Masses)+endx])
yFit = fit2[0]*xFit + fit2[1]
ax.plot(xFit, yFit, "-b", zorder=-1)
ax.set_xlabel("M(T) GeV/$c^{2}$")
ax.set_ylabel("$\sigma$ from fitted M(5j)")
ax.xaxis.label.set_fontsize(18)
ax.yaxis.label.set_fontsize(18)
ax.xaxis.set_label_coords(0.84, -0.07)
ax.yaxis.set_label_coords(-0.07, 0.67)
ax.set_xlim(left=Masses[0]*0.95,right=Masses[-1]*1.05)
ax.set_ylim(bottom=min(SigmaValues)*0.5,top=max(SigmaValues)*1.5)
#plt.legend(loc='lower right')
fig.savefig("Fitted_TpSigma_V8.pdf")
fig.savefig("Fitted_TpSigma_V8.png")

print "Masses and Widths for each mass point from linear fit:"
M=fit1[0]*np.array(Masses)+fit1[1]
W=fit2[0]*np.array(Masses)+fit2[1]
Nsigmas=1.0
W=Nsigmas*np.array(W)
for i in xrange(len(Masses)):
    print "M={0:0.2f}, Width={1:0.2f}".format(M[i], W[i])
    
ROOT.gDirectory.Delete("*CutF*")
```
