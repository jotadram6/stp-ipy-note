```python
Version="/home/jruizalv/V1/"

#Cuts
HTCut = ROOT.TCut("THT>600")
J1PTCut = ROOT.TCut("jet1_pt>150")
L3 = ROOT.TCut("Number_CSVLbtagged_jets>=3")
M3 = ROOT.TCut("Number_CSVMbtagged_jets>=3")
M2L1 = ROOT.TCut("Number_CSVMbtagged_jets>=2 && Number_CSVLbtagged_jets>=1")
M1L2 = ROOT.TCut("Number_CSVMbtagged_jets>=1 && Number_CSVLbtagged_jets>=2")
M2T1 = ROOT.TCut("Number_CSVMbtagged_jets>=2 && Number_CSVTbtagged_jets>=1")
M1T2 = ROOT.TCut("Number_CSVMbtagged_jets>=1 && Number_CSVTbtagged_jets>=2")
T3 = ROOT.TCut("Number_CSVTbtagged_jets>=3")
T2L1 = ROOT.TCut("Number_CSVTbtagged_jets>=2 && Number_CSVLbtagged_jets>=1")
T1L2 = ROOT.TCut("Number_CSVTbtagged_jets>=1 && Number_CSVLbtagged_jets>=2")
T1M1L1 = ROOT.TCut("Number_CSVTbtagged_jets>=1 && Number_CSVMbtagged_jets>=1 && Number_CSVLbtagged_jets>=1")

AllSamples=[]
TotalEntries=[]

for f in files[-1:]:
    CutsChain=ROOT.TChain("cuts")
    CutsChain.Add(Version+f)
    entries = CutsChain.GetEntries()
    if files.index(f)>=0 and files.index(f)<=NsignalFiles-1: print "Filling signal..."
    elif files.index(f)>0: print "Filling bkgs..."
    print "Entries for file ", f, " are: ", entries
    TempTree= ROOT.TNtuple(ListTreesNamesA[files.index(f)],ListTreesNamesA[files.index(f)],TreeStructure)
    TC=0; L3C=0; M3C=0; M2L1C=0; M1L2C=0; M2T1C=0; M1T2C=0; T3C=0; T2L1C=0; T1L2C=0; T1M1L1C=0
    for i in xrange(entries):
        CutsChain.GetEntry(i)
        if CutsChain.THT<600: continue
        if CutsChain.jet1_pt<150: continue
        TC+=1
        if (CutsChain.Number_CSVLbtagged_jets>=3): L3C+=1
        if (CutsChain.Number_CSVMbtagged_jets>=3): M3C+=1
        if (CutsChain.Number_CSVMbtagged_jets>=2 and CutsChain.Number_CSVLbtagged_jets>=1): M2L1C+=1
        if (CutsChain.Number_CSVMbtagged_jets>=1 and CutsChain.Number_CSVLbtagged_jets>=2): M1L2C+=1
        if (CutsChain.Number_CSVMbtagged_jets>=2 and CutsChain.Number_CSVTbtagged_jets>=1): M2T1C+=1
        if (CutsChain.Number_CSVMbtagged_jets>=1 and CutsChain.Number_CSVTbtagged_jets>=2): M1T2C+=1
        if (CutsChain.Number_CSVTbtagged_jets>=3): T3C+=1
        if (CutsChain.Number_CSVTbtagged_jets>=2 and CutsChain.Number_CSVLbtagged_jets>=1): T2L1C+=1
        if (CutsChain.Number_CSVTbtagged_jets>=1 and CutsChain.Number_CSVLbtagged_jets>=2): T1L2C+=1
        if (CutsChain.Number_CSVTbtagged_jets>=1 and CutsChain.Number_CSVMbtagged_jets>=1 and CutsChain.Number_CSVLbtagged_jets>=1): T1M1L1C+=1
    TotalEntries.append(TC)
    AllSamples.append([L3C, M3C, M2L1C, M1L2C, M2T1C, M1T2C, T3C, T2L1C, T1L2C, T1M1L1C])
```
